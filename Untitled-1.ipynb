```python
import torch
import torch.nn as nn
import scanpy as sc
import torch.nn.functional as F

adata = sc.read_h5ad('human_tongue.h5ad')
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')

cell_types = list(set(adata.obs['cell_type'].tolist()))
idx2type = {i: cell_type for i, cell_type in enumerate(cell_types)}
type2idx = {cell_type: i for i, cell_type in enumerate(cell_types)}

xs = torch.tensor(adata.X.toarray(), dtype=torch.float32, device=device)
ys = [type2idx[cell_type] for cell_type in adata.obs['cell_type'].tolist()]
ys = torch.tensor(ys, dtype=torch.long, device=device)
plt.plot(range(total_steps), losses)



```

```python
class LogisticRegression(torch.nn.Module):
    def __init__(self, input_dim, output_dim):
        super(LogisticRegression, self).__init__()  # noqa: UP008
        self.linear = torch.nn.Linear(input_dim, output_dim)

    def forward(self, x):
        outputs = self.linear(x)
        return outputs
```

```python
import matplotlib.pyplot as plt

def train_model(model, optimizer, n_epochs, train_ysxs):
    model.train()

    losses = []
    total_steps = 0
    for epoch in range(n_epochs):
        train_ysxs = train_ysxs[torch.randperm(train_ysxs.size(0))]
        for i in range(0, train_ysxs.shape[0], batch_size):
            x_batch = train_ysxs[i:i+batch_size, 1:]
            y_batch = train_ysxs[i:i+batch_size, 0]
            optimizer.zero_grad()
            out = model(x_batch)
            y_batch = torch.tensor(y_batch, dtype=torch.long, device=device)
            loss = F.cross_entropy(out, y_batch)
            
            losses.append(loss.item())
            if len(losses) >= 10 and epoch == n_epochs-1 and i >= train_ysxs.shape[0] - 10*batch_size:
                print(f'Epoch {epoch}, batch {i}, loss: {loss.item():.4f}')

            loss.backward()
            optimizer.step()
            total_steps += 1

    plt.figure(figsize=(5, 3))
    plt.plot(range(total_steps), losses)
    plt.xlabel('Training Step')
    plt.ylabel('Loss')
    plt.title('Training Loss Over Time')
    plt.grid(True)
    plt.show()
```

```python
import matplotlib.pyplot as plt

def train_sampling_model(model, n_samples, optimizer, n_epochs, train_ysxs):
    model.train()

    losses = []
    total_steps = 0
    for epoch in range(n_epochs):
        train_ysxs = train_ysxs[torch.randperm(train_ysxs.size(0))]
        for i in range(0, train_ysxs.shape[0], batch_size):
            x_batch = train_ysxs[i:i+batch_size, 1:]
            y_batch = train_ysxs[i:i+batch_size, 0]
            x_samples = torch.multinomial(x_batch, n_samples, replacement=True)
            x_batch = torch.zeros_like(x_batch)
            x_batch.scatter_(1, x_samples, 1, reduce='add')
            
            
            optimizer.zero_grad()
            out = model(x_batch)
            y_batch = torch.tensor(y_batch, dtype=torch.long, device=device)
            loss = F.cross_entropy(out, y_batch)
            
            losses.append(loss.item())
            if len(losses) >= 10 and epoch == n_epochs-1 and i >= train_ysxs.shape[0] - 10*batch_size:
                print(f'Epoch {epoch}, batch {i}, loss: {loss.item():.4f}')

            loss.backward()
            optimizer.step()
            total_steps += 1

    plt.figure(figsize=(5, 3))
    plt.plot(range(total_steps), losses)
    plt.xlabel('Training Step')
    plt.ylabel('Loss')
    plt.title('Training Loss Over Time')
    plt.grid(True)
    plt.show()
```

```python
def evaluate_model(model, test_ysxs, batch_size, device, idx2type):
    losses = []
    model.eval()
    test_ysxs = test_ysxs[torch.randperm(test_ysxs.size(0))]
    for i in range(10):
        y_batch = test_ysxs[i*batch_size:(i+1)*batch_size, 0]
        x_batch = test_ysxs[i*batch_size:(i+1)*batch_size, 1:]
        out = model(x_batch)
        probs = torch.softmax(out, dim=1)
        pred = torch.argmax(probs, dim=1)
        y_batch = torch.tensor(y_batch, dtype=torch.long, device=device)
        loss = F.cross_entropy(out, y_batch)
        losses.append(loss.item())
        predictions = []
        for i in range(len(pred)):
            predictions.append((idx2type[pred[i].item()], idx2type[y_batch[i].item()]))

    print(f'test loss: {torch.tensor(losses).mean().item():.4f}, std: {torch.tensor(losses).std().item():.4f}')
    for pred_type, true_type in predictions[:10]:
        print(pred_type, '        ', true_type)
```

```python
def evaluate_sampling_model(model, n_samples, test_ysxs, batch_size, device, idx2type):
    losses = []
    model.eval()
    test_ysxs = test_ysxs[torch.randperm(test_ysxs.size(0))]
    for i in range(10):
        y_batch = test_ysxs[i*batch_size:(i+1)*batch_size, 0]
        x_batch = test_ysxs[i*batch_size:(i+1)*batch_size, 1:]
        x_samples = torch.multinomial(x_batch, n_samples, replacement=True)
        x_batch = torch.zeros_like(x_batch)
        x_batch.scatter_(1, x_samples, 1, reduce='add')
        out = model(x_batch)
        probs = torch.softmax(out, dim=1)
        pred = torch.argmax(probs, dim=1)
        y_batch = torch.tensor(y_batch, dtype=torch.long, device=device)
        loss = F.cross_entropy(out, y_batch)
        losses.append(loss.item())
        predictions = []
        for i in range(len(pred)):
            predictions.append((idx2type[pred[i].item()], idx2type[y_batch[i].item()]))

    print(f'test loss: {torch.tensor(losses).mean().item():.4f}, std: {torch.tensor(losses).std().item():.4f}')
    for pred_type, true_type in predictions[:10]:
        print(pred_type, '        ', true_type)
```
