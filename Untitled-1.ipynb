```python
import torch
import torch.nn as nn
import scanpy as sc
import torch.nn.functional as F

adata = sc.read_h5ad('human_tongue.h5ad')
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')

cell_types = list(set(adata.obs['cell_type'].tolist()))
idx2type = {i: cell_type for i, cell_type in enumerate(cell_types)}
type2idx = {cell_type: i for i, cell_type in enumerate(cell_types)}

xs = torch.tensor(adata.X.toarray(), dtype=torch.float32, device=device)
ys = [type2idx[cell_type] for cell_type in adata.obs['cell_type'].tolist()]
ys = torch.tensor(ys, dtype=torch.long, device=device)



```

```python
class LogisticRegression(torch.nn.Module):
    def __init__(self, input_dim, output_dim):
        super(LogisticRegression, self).__init__()  # noqa: UP008
        self.linear = torch.nn.Linear(input_dim, output_dim)

    def forward(self, x):
        outputs = self.linear(x)
        return outputs
```

```python
model = LogisticRegression(xs.shape[1], len(cell_types)).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-05)

batch_size = 64
n_epochs = 100
```

```python
model = LogisticRegression(xs.shape[1], len(cell_types)).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-05)

model.train()
ysxs = torch.cat([ys.unsqueeze(1), xs], dim=1)
losses = []
total_steps = 0
for epoch in range(n_epochs):
    ysxs = ysxs[torch.randperm(ysxs.size(0))]
    for i in range(0, ysxs.shape[0], batch_size):
        x_batch = ysxs[i:i+batch_size, 1:]
        y_batch = ysxs[i:i+batch_size, 0]
        optimizer.zero_grad()
        out = model(x_batch)
        y_batch = torch.tensor(y_batch, dtype=torch.long, device=device)
        loss = F.cross_entropy(out, y_batch)
        
        losses.append(loss.item())
        if i % 100 == 0:
            print(f'Epoch {epoch}, batch {i}, loss: {loss.item():.4f}')

        loss.backward()
        optimizer.step()
        total_steps += 1

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(range(total_steps), losses)
plt.xlabel('Training Step')
plt.ylabel('Loss')
plt.title('Training Loss Over Time')
plt.grid(True)
plt.show()
```

```python
model.eval()
ysxs = torch.cat([ys.unsqueeze(1), xs], dim=1)
ysxs = ysxs[torch.randperm(ysxs.size(0))]
for i in range(100):
    y_batch = ysxs[i*batch_size:(i+1)*batch_size, 0]
    x_batch = ysxs[i*batch_size:(i+1)*batch_size, 1:]
    out = model(x_batch)
    probs = torch.softmax(out, dim=1)
    pred = torch.argmax(probs, dim=1)
    print(f'batch {i}')
    for i in range(len(pred)):
        print(idx2type[pred[i].item()], '        ', idx2type[y_batch[i].item()])
```
