```python
import anndata as ad
import torch
import scanpy as sc
```

```python
adata = sc.read_h5ad('../human_tongue.h5ad')
```

```python
sc.pp.neighbors(adata)
sc.tl.umap(adata)
sc.pl.umap(adata, color='cell_type')
```

```python
t0_embs = torch.load('cell_emb.pt')
t0_adata = ad.AnnData(t0_embs.numpy())
t0_adata.obs = adata.obs
```

```python
sc.pp.neighbors(t0_adata)
sc.tl.umap(t0_adata)
sc.pl.umap(t0_adata, color='cell_type')
```

```python
e1_embs = torch.load('../cell_emb_1_epoch.pt')
e1_adata = ad.AnnData(e1_embs.detach().numpy())
e1_adata.obs = adata.obs
```

```python
#embeddings of human tongue after one epoch
sc.pp.neighbors(e1_adata)
sc.tl.umap(e1_adata)
sc.pl.umap(e1_adata, color='cell_type')
```
