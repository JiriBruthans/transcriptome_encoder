```python
import pandas as pd
import matplotlib.pyplot as plt

with open('losses.txt', 'r') as f:
    lines = f.readlines()

steps = []
losses = []
for line in lines:
    # Format: "loss: 0.7042100429534912 for step 0"
    parts = line.split()
    steps.append(int(parts[-1]))
    losses.append(float(parts[1]))

df = pd.DataFrame({'step': steps, 'loss': losses})

plt.figure(figsize=(10, 6))
plt.plot(df['step'], df['loss'], marker='o')
plt.title('Training Loss Over Time')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

with open('losses.txt', 'r') as f:
    lines = f.readlines()

steps = []
losses = []
for line in lines:
    parts = line.split()
    steps.append(int(parts[-1]))
    losses.append(float(parts[1]))

df = pd.DataFrame({'step': steps, 'loss': losses})
df['log_step'] = np.log(df['step'] + 1)  # Add 1 to handle step 0

plt.figure(figsize=(10, 6))
plt.plot(df['log_step'], df['loss'], marker='o')
plt.title('Training Loss vs Log Steps')
plt.xlabel('Log Step')
plt.ylabel('Loss')
plt.grid(True)
plt.show()
```
